```python
import json
import pandas as pd
```

```python
folder = "../words/"

ch = "_Language.chinese"
en = "_Language.english"
k_json_file = ".k.json"
unk_json_file = ".unk.json"

train_ch_k_json_file = "train" + ch + k_json_file
train_ch_unk_json_file = "train" + ch + unk_json_file
dev_ch_k_json_file = "dev" + ch + k_json_file 
dev_ch_unk_json_file = "dev" + ch + unk_json_file
test_ch_k_json_file = "test" + ch + k_json_file
test_ch_unk_json_file = "test" + ch + unk_json_file

train_en_k_json_file = "train" + en + k_json_file
train_en_unk_json_file = "train" + en + unk_json_file
dev_en_k_json_file = "dev" + en + k_json_file 
dev_en_unk_json_file = "dev" + en + unk_json_file
test_en_k_json_file = "test" + en + k_json_file
test_en_unk_json_file = "test" + en + unk_json_file
```

```python
with open(folder + train_ch_unk_json_file, 'r', encoding='utf-8') as fp1:
    train_ch_unk_json = json.load(fp1)
with open(folder + test_ch_unk_json_file, 'r', encoding='utf-8') as fp2:
    test_ch_unk_json = json.load(fp2)
```

```python
df_train_ch_unk = pd.DataFrame.from_dict(train_ch_unk_json, orient='index')
df_test_ch_unk = pd.DataFrame.from_dict(test_ch_unk_json, orient='index')
```

```python
df_train_ch_unk = df_train_ch_unk.sort_values(by = 0, ascending=False)
df_test_ch_unk = df_test_ch_unk.sort_values(by = 0, ascending=False)
```
